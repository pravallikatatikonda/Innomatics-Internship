```python
# Problem 1: Social Media – Post Engagement Analyzer
def analyze_post_engagement(likes_list):
    total_likes = 0

    for likes in likes_list:
        total_likes += likes

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    return total_likes, status


# Example Usage
likes_data = [200, 300, 150, 400]
total, result = analyze_post_engagement(likes_data)

print("Total Likes:", total)
print("Post Status:", result)
```

```python
# Problem 3: Agriculture – Rainfall Adequacy Checker
def rainfall_checker(rainfall_data, required_level):
    total = 0

    for rain in rainfall_data:
        total += rain

    average = total / len(rainfall_data)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    return average, status


# Example Usage
rainfall = [70, 75, 80, 65]
required = 70
avg, result = rainfall_checker(rainfall, required)

print("Average Rainfall:", avg)
print("Rainfall Status:", result)
```

```python
# Problem 4: Social Media – Duplicate Account Detection
def detect_duplicates(usernames):
    unique_users = set(usernames)

    if len(unique_users) < len(usernames):
        return "Yes"
    else:
        return "No"


# Example Usage
users = ["john", "alice", "john", "mike"]
print("Duplicate Accounts Found:", detect_duplicates(users))
```

```python
# Problem 6: Agriculture – Premium Crop Price Filter
def filter_premium_crops(prices):
    premium = []

    for price in prices:
        if price > 2000:
            premium.append(price)

    return premium


# Example Usage
crop_prices = [1500, 2500, 1800, 3200]
print("Premium Crops:", filter_premium_crops(crop_prices))
```

```python
# Problem 7: System Monitoring – Application Health Checker
def application_health_checker(error_count):
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    return status


# Example Usage
errors = 7
print("Error Count:", errors)
print("System Status:", application_health_checker(errors))
```

```python
# Problem 8: Banking – Daily Transaction Limit Checker
def transaction_checker(amount):
    daily_limit = 50000

    if amount <= daily_limit:
        return "Approved"
    else:
        return "Rejected"


# Example Usage
amount = 60000
print("Transaction Amount:", amount)
print("Transaction Status:", transaction_checker(amount))
```

```python
# Problem 9: E-Learning – Student Attendance Eligibility System
def attendance_eligibility(attendance_list):
    total_classes = len(attendance_list)
    attended = 0

    for status in attendance_list:
        if status == "Present":
            attended += 1

    percentage = (attended / total_classes) * 100

    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    return percentage, eligibility


# Example Usage
attendance = ["Present", "Present", "Absent", "Present", "Present"]
percent, result = attendance_eligibility(attendance)

print("Attendance Percentage:", percent)
print("Exam Eligibility:", result)
```

```python
# Problem 10: Smart Electricity Bill Analyzer
def electricity_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage Classification
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage


# Example Usage
units = 250
bill_amount, usage_status = electricity_bill(units)

print("Total Bill:", bill_amount)
print("Usage Status:", usage_status)
```
