```python
# Problem 1: Smart Parking Lot Management
def parking_lot_manager(capacity, logs):
    count = 0

    for log in logs:
        if log == "IN":
            count += 1
        elif log == "OUT":
            count -= 1

    status = "Full" if count >= capacity else "Available"

    print("Currently Parked Vehicles:", count)
    print("Parking Status:", status)


# Sample Input
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

parking_lot_manager(capacity, logs)
```

```python
#Problem 2: Online Food Delivery Time Estimator
def delivery_time(distance, traffic, weather):
    time = distance * 5

    if traffic == "High":
        time += 15
    elif traffic == "Medium":
        time += 8


    if weather == "Stormy":
        time += 20

    print("Estimated Delivery Time:", time, "minutes")


# Sample Input
delivery_time(8, "High", "Rainy")
```

```python
# Problem 3: Movie Theatre Seat Occupancy Analyzer
def seat_occupancy(total_seats, booked_seats):
    booked = len(booked_seats)

    occupancy = (booked / total_seats) * 100

    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 70:
        status = "Almost Full"
    else:
        status = "Available"

    print("Occupancy:", int(occupancy), "%")
    print("Show Status:", status)


# Sample Input
total = 200
booked = [1] * 150

seat_occupancy(total, booked)
```

```python
# Problem 4: Cloud Server Load Classification
def server_load(cpu_readings):
    avg = sum(cpu_readings) / len(cpu_readings)

    if avg < 50:
        status = "Normal"
    elif avg <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Average CPU Load:", int(avg), "%")
    print("Server Status:", status)


# Sample Input
readings = [45, 60, 70, 85, 90]

server_load(readings)
```

```python
# Problem 5: Smart Classroom Resource Monitor
def resource_monitor(usage):
    overused = []

    for resource, hours in usage.items():
        if hours > 7:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
resources = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

resource_monitor(resources)
```

```python
# Problem 6: Online Event Registration Controller
def registration_controller(capacity, registrations):

    if registrations <= capacity:
        confirmed = registrations
        waitlist = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)


# Sample Input
registration_controller(100, 105)
```
