```python
#Problem 1: Mobile Recharge Validation System
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("Recharge amount must be ≥ ₹50")
        return False

    if amount not in valid_plans:
        print("Invalid plan selected")
        return False

    return True


def recharge_system():
    while True:
        try:
            amount = int(input("Enter recharge amount: "))

            if validate_recharge(amount):
                print("Recharge Successful ✅")
                break
            else:
                print("Please try again.\n")

        except ValueError:
            print("Invalid input. Enter numeric value.\n")


# Call function
recharge_system()
```

```python
# Problem 2: Inventory Reorder Alert System
def inventory_alert(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"{product} → Reorder Alert")
        else:
            print(f"{product} → Stock OK")


def run_inventory_system():
    products = {
        "Laptop": 10,
        "Mouse": 25,
        "Keyboard": 12,
        "Monitor": 18
    }

    inventory_alert(products)


# Call function
run_inventory_system()
```

```python
# Problem 3: Student Result Processing System
def calculate_average(marks):
    total = 0

    for mark in marks:
        total += mark

    average = total / len(marks)
    return average


def student_result_system():
    marks = []
    n = int(input("Enter number of subjects: "))

    for i in range(n):
        mark = float(input(f"Enter marks for subject {i+1}: "))
        marks.append(mark)

    avg = calculate_average(marks)
    print("Average Marks:", avg)

    if avg >= 50:
        print("Result: Pass")
    else:
        print("Result: Fail")


# Call function
student_result_system()
```

```python
# Problem 4: Cab Fare Estimator with Retry Option
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12

    total = base_fare + (distance * per_km_rate)

    if peak.lower() == "yes":
        total += total * 0.25

    return total


def cab_fare_system():
    while True:
        try:
            distance = float(input("Enter distance (km): "))
            peak = input("Peak hour? (yes/no): ")

            fare = calculate_fare(distance, peak)
            print("Total Fare: ₹", fare)
            break

        except ValueError:
            print("Invalid input. Try again.\n")


# Call function
cab_fare_system()
```

```python
# Problem 5: Employee Attendance Eligibility Checker
def calculate_attendance(attendance_list):
    present = 0

    for status in attendance_list:
        if status == "P":
            present += 1

    percentage = (present / len(attendance_list)) * 100
    return percentage


def attendance_system():
    attendance = []
    days = int(input("Enter total working days: "))

    for i in range(days):
        status = input(f"Day {i+1} (P/A): ").upper()
        attendance.append(status)

    percentage = calculate_attendance(attendance)
    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        print("Status: Eligible")
    else:
        print("Status: Not Eligible")


# Call function
attendance_system()
```

```python
# Problem 6: Password Strength Checker
def check_password(password):
    special_chars = "@#$"
    has_digit = False
    has_special = False

    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password ✅"
    else:
        return "Weak Password (Must contain at least one digit and one special character)"


def password_system():
    password = input("Enter password: ")
    result = check_password(password)
    print(result)


# Call function
password_system()
```
