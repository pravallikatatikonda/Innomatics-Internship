```python
# Problem Statement 1: Employee Performance Bonus Eligibility
# Employee data
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}

#highest score
max_score=max(employees.values())

#find top performers
top_performers=[name for name,score in employees.items() if score==max_score]

#Display result
print("Top Performers Eligible for Bonus:",", ".join(top_performers),f"(Score: {max_score})")
```

```python
# Problem Statement 2: Search Query Keyword Analysis
#Input Sentence
sentence="Buy mobile phone buy phone online"

# Convert to lowercase
sentence=sentence.lower()

#Remove Punctuation
import string
for char in string.punctuation:
  sentence=sentence.replace(char," ")

# Split into words
words=sentence.split()

# Count frequency
frequency={}
for word in words:
  frequency[word]=frequency.get(word,0)+1

# Display words with frequency > 1
result = {word: count for word, count in frequency.items() if count > 1}

print(result)
```

```python
# Problem Statement 3: Sensor Data Validation

# Sensor readings
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

# Check each reading
for index, value in enumerate(sensor_readings):
    if value % 2 == 0:   # Even number
        valid_readings.append((index, value))

# Display result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# Problem Statement 4: Email Domain Usage Analysis

# Email list
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# Count domains
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

total = len(emails)

# Calculate percentage
for domain, count in domain_count.items():
    percentage = (count / total) * 100
    print(f"{domain}: {int(percentage)}%")
```

```python
# Problem Statement 5: Sales Spike Detection

sales = [1200, 1500, 900, 2200, 1400, 3000]

average = sum(sales) / len(sales)

threshold = average * 1.25

for i, value in enumerate(sales):
    if value >= threshold:
        print(f"Day {i+1}: {value}")
```

```python
# Problem Statement 6: Duplicate User ID Detection

# User IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

count_ids = {}

# Count IDs
for user in user_ids:
    count_ids[user] = count_ids.get(user, 0) + 1

# Display duplicates
for user, count in count_ids.items():
    if count > 1:
        print(f"{user} → {count} times")
```
