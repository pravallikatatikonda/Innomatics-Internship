```python

# Problem 1: Smart Light Controller
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def display_status(self):
        print(f"{self.name} is {self.status}")


# Input
name = input("Light Name: ")
action = input("Action: ")

light = SmartLight(name)

if action == "ON":
    light.turn_on()
else:
    light.turn_off()

light.display_status()
```

```python
# Problem 2: Employee ID Card System
class Employee:
    def __init__(self, name, emp_id, dept):
        self.name = name
        self.emp_id = emp_id
        self.dept = dept

    def display(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.dept)


# Input
name = input("Employee Name: ")
emp_id = input("Employee ID: ")
dept = input("Department: ")

emp = Employee(name, emp_id, dept)
emp.display()
```

```python
# Problem 3: Mobile Contact Record
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# Input
name = input("Contact Name: ")
phone = input("Phone Number: ")

contact = Contact(name, phone)
contact.display()
```

```python
# Problem 4: Product Price Tag Generator
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))


# Input
name = input("Product Name: ")
price = input("Price: ")

product = Product(name, price)
product.display()
```

```python
# Problem 5: Movie Rating Display System
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")


# Input
name = input("Movie Name: ")
rating = input("Rating: ")

movie = Movie(name, rating)
movie.display()
```

```python
# Problem 6: Delivery Address Manager
class Delivery:
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address

    def display(self):
        print("Delivery Details")
        print("Customer:", self.customer)
        print("Address:", self.address)


# Input
name = input("Customer Name: ")
address = input("Address: ")

delivery = Delivery(name, address)
delivery.display()
```
